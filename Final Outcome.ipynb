```python
import pandas as pd
import os
import warnings
from datetime import datetime
warnings.filterwarnings(action='ignore')
```

```python
data_ubereats = pd.read_csv(os.path.join('final_uber.csv'))
data_grubhub = pd.read_csv(os.path.join('Downloads','grubhub.csv'))
```

```python
ueats_df = data_ubereats.head(80)
ghub_df = data_grubhub.head(80)
```

```python
merged_df = pd.merge(ghub_df, ueats_df, on=['b_name'])
merged_df_2 = pd.merge(ghub_df, ueats_df, on=['vb_name'])
```

```python
merged_df = merged_df.append(merged_df_2)
```

```python
merged_df['starttime'] = pd.to_datetime(merged_df['starttime'])
merged_df['endtime'] = pd.to_datetime(merged_df['endtime'])
merged_df['ueats_end']= pd.to_datetime(merged_df['ueats_end'])
merged_df['ueats_start'] =  pd.to_datetime(merged_df['ueats_start'])
```

```python
merged_df['Ubereats Business Hours'] = merged_df['ueats_end'] - merged_df['ueats_start'] 
merged_df['Grubhub Business Hours'] = merged_df['endtime']-merged_df['starttime']
```

```python
from datetime import datetime, timedelta

def compare_time(row):
    time_diff = row['Ubereats Business Hours'] - row['Grubhub Business Hours']
    if time_diff < timedelta(minutes=5):
        return 'Out of range'
    elif time_diff > timedelta(minutes=5):
        return 'In Range'
    elif time_diff <= timedelta(minutes=5):
        return 'Out of Range with 5 mins difference'
    else:
        return 'Out of Range'

```

```python
merged_df['is_out_of_range'] = merged_df.apply(compare_time, axis=1)
df= merged_df[['slug_x','slug_y','Grubhub Business Hours','Ubereats Business Hours','is_out_of_range']]
df
```

```python
df['Ubereats Business Hours']= df['Ubereats Business Hours'].apply(lambda x: str(x)[1:].split()[-1])
df['Grubhub Business Hours']= df['Grubhub Business Hours'].apply(lambda x: str(x)[1:].split()[-1])
df = df.rename(columns={'slug_x': 'Grubhub slug'})
df = df.rename(columns={'slug_y': 'Uber Eats slug'})
```

```python
df = df[['Grubhub slug','Grubhub Business Hours','Uber Eats slug','Ubereats Business Hours','is_out_of_range']]
```
