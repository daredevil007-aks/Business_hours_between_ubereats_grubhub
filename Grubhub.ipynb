```python
data = pd.read_csv(os.path.join('grubhub.csv'))
df = data.head(100)
looping = df['response']
```

```python
endtime_final = []

for i in looping:
    time.sleep(3)
    parsed_data = json.loads(i)
    try:
        time.sleep(3)
        endtime = parsed_data['availability_by_catalog']['STANDARD_DELIVERY']['schedule_rules'][0]['to']
        endtime_final.append(endtime)
        print(endtime_final)
        
    except Exception as e:
        print(e)
        time.sleep(3)
        endtime_final.append("not time available")
        pass
        
```

```python
import json
import time

starttime_final =[]

for i in looping:
    time.sleep(3)
    parsed_data = json.loads(i)
    try:
        time.sleep(3)
        starttime = parsed_data['availability_by_catalog']['STANDARD_DELIVERY']['schedule_rules'][0]['from']
        starttime_final.append(starttime)
        print(starttime_final)
        
    except Exception as e:
        print(e)
        time.sleep(3)
        starttime_final.append("not time available")
        pass
```

```python
df['starttime'] = pd.DataFrame(starttime_final)
df['endtime']=pd.DataFrame(endtime_final)
df['starttime'] = df['starttime'].replace('not time available', pd.NA)
df.dropna(subset=['starttime'], inplace=True)
```

```python
df.to_csv('grubhub.csv')
```
